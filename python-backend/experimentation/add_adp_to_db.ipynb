```python
half_ppr_df = pd.read_csv("../FP_2025_Half_PPR_ADP.csv")
ppr_df = pd.read_csv("../FP_2025_PPR_ADP.csv")
std_df = pd.read_csv("../FP_2025_STD_ADP.csv")

half_ppr_df.head()
```

```python
# load in database

# --- Supabase connection ---
url="https://xolsmdlhmwavftlszhml.supabase.co"
key="eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJpc3MiOiJzdXBhYmFzZSIsInJlZiI6InhvbHNtZGxobXdhdmZ0bHN6aG1sIiwicm9sZSI6ImFub24iLCJpYXQiOjE3NTQ1OTg2ODgsImV4cCI6MjA3MDE3NDY4OH0.Ds4ymqFyyRfYVfdOZDEGWThV4jd5bsDLCsLB--6bQYQ"

supabase: Client = create_client(url, key)

# Get all players from Supabase
response = supabase.table("nfl_players_2025").select("*").execute()
players = response.data
```

```python
for player in players:
    full_name = f"{player['first_name']} {player['last_name']}"

    # Half PPR
    match_half = half_ppr_df[half_ppr_df["Player"] == full_name]
    if not match_half.empty:
        adp_half = float(match_half.iloc[0]["AVG"])
    else:
        adp_half = 1000

    # PPR
    match_ppr = ppr_df[ppr_df["Player"] == full_name]
    if not match_ppr.empty:
        adp_ppr = float(match_ppr.iloc[0]["AVG"])
    else:
        adp_ppr = 1000

    # STD
    match_std = std_df[std_df["Player"] == full_name]
    if not match_std.empty:
        adp_std = float(match_std.iloc[0]["AVG"])
    else:
        adp_std = 1000

    # Update player record in Supabase
    supabase.table("nfl_players_2025").update({
        "fp_adp_half_ppr_avg": adp_half,
        "fp_adp_ppr_avg": adp_ppr,
        "fp_adp_std_avg": adp_std
    }).eq("id", player["id"]).execute()

    if adp_half == 1000 and adp_ppr == 1000 and adp_std == 1000:
        print(f"Could not update ADP for {full_name}")

print("Done.")
```
