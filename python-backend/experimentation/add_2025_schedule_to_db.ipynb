```python
import requests
import pandas as pd

# url = "https://site.api.espn.com/apis/site/v2/sports/football/nfl/scoreboard?year=2024&seasontype=2&week=1"
# data = requests.get(url).json()
# You can parse data['events'] or merge based on 'schedule' depending on structure.


games = []
year = 2024
for week in range(1, 19):
    url = f"https://site.api.espn.com/apis/site/v2/sports/football/nfl/scoreboard?dates=2024&seasontype=2&week={week}"
    data = requests.get(url).json()
    for game in data['events']:
        home_team_abbrev = game['competitions'][0]['competitors'][0]['team']['abbreviation']
        away_team_abbrev = game['competitions'][0]['competitors'][1]['team']['abbreviation']
        
        if away_team_abbrev == "LAR":
            away_team_abbrev = "LA"
        if home_team_abbrev == "LAR":
            home_team_abbrev = "LA"
        
        if away_team_abbrev == "WSH":
            away_team_abbrev = "WAS"
        if home_team_abbrev == "WSH":
            home_team_abbrev = "WAS"
                
        home_team_id = team_abbrev_to_id[home_team_abbrev]
        away_team_id = team_abbrev_to_id[away_team_abbrev]
        
        games.append({
            "week": week,
            "year": year,
            "home_team_id": home_team_id,
            "away_team_id": away_team_id
        })
        
```

```python
supabase.table("nfl_schedule").insert(games).execute()
```
